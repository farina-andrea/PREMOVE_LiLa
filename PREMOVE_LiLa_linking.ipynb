# PREMOVE: linking to LiLa

## WordNet synsets: attestations

This script takes a list of WordNet 3.1 token URIs from `PREMOVE_WN_3.csv` and retrieves their corresponding passage URIs from the LiLa Knowledge Base via SPARQL queries.

````python
import pandas as pd
from SPARQLWrapper import SPARQLWrapper, JSON, POST
import time

# 1. Load CSV
df = pd.read_csv("PREMOVE_WN_3.1.csv")
df["URIs"] = df["URIs"].astype(str).str.strip()

# 2. Setup SPARQL endpoint
sparql = SPARQLWrapper("https://lila-erc.eu/sparql/lila_knowledge_base/sparql")
sparql.setMethod(POST)
sparql.setReturnFormat(JSON)

# 3. Function to query a batch of URIs
def query_batch(uris):
    values_block = " ".join(f"<{uri}>" for uri in uris)
    query = f"""
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    PREFIX lila: <http://lila-erc.eu/ontologies/lila/>
    PREFIX dc: <http://purl.org/dc/elements/1.1/>
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
    PREFIX powla: <http://purl.org/powla/powla.owl#>

    SELECT ?token_URI ?passage_URI
    WHERE {{
      VALUES ?token_URI {{ {values_block} }}
      ?passage_URI powla:hasChild ?token_URI .
    }}
    """
    sparql.setQuery(query)
    results = sparql.query().convert()
    return [
        (res["token_URI"]["value"], res["passage_URI"]["value"])
        for res in results["results"]["bindings"]
    ]

# 4. Loop through all 960 URIs in batches of 50
batch_size = 50
all_results = []
for i in range(0, len(df), batch_size):
    batch = df["URIs"].iloc[i:i+batch_size].tolist()
    print(f"üîé Querying batch {i // batch_size + 1}/{-(-len(df)//batch_size)} ({len(batch)} URIs)...")
    try:
        results = query_batch(batch)
        all_results.extend(results)
    except Exception as e:
        print(f"‚ö†Ô∏è Error in batch {i // batch_size + 1}: {e}")
    time.sleep(0.2)  # small delay to avoid stressing the server

# 5. Save results
output_df = pd.DataFrame(all_results, columns=["URIs", "passage_URIs"])
output_df.to_csv("tokens_and_passages_URIs.csv", index=False)

print("‚úÖ Finished! Results saved to tokens_and_passages_URIs.csv")
````

I filter out passages belonging to the SentenceLayer from the SPARQL query results, keeping only higher-level passage URIs.

````python
# 5a. Filter out SentenceLayer passages
filtered_results = [
    (token, passage) for token, passage in all_results
    if "/Sentence" not in passage
]

# 5b. Save filtered results
output_df = pd.DataFrame(filtered_results, columns=["URIs", "passage_URIs"])
output_df.to_csv("tokens_and_passages_URIs_filtered.csv", index=False)

print("‚úÖ Finished! Filtered results saved to tokens_and_passages_URIs_filtered.csv")
````

I then merge these filtered passages with the original token‚Äìsynset mapping, producing a final CSV file that links each token URI to both its synset and its passage URI.

````python
# Load the filtered passages CSV
passages_df = pd.read_csv("tokens_and_passages_URIs_filtered.csv")

# Load your original token CSV
tokens_df = pd.read_csv("PREMOVE_WN_3.1.csv")
tokens_df["URIs"] = tokens_df["URIs"].astype(str).str.strip()  # ensure no whitespace issues

# Merge on 'URIs'
merged_df = pd.merge(tokens_df, passages_df, on="URIs", how="left")

# Save the merged CSV
merged_df.to_csv("token_passage_synset_URIs.csv", index=False)

print("‚úÖ Merged CSV saved as token_passage_synset_URIs.csv")
````

As an example, I create an RDF graph for one Latin lexical concept and its attestation, then visualizes the data as a network. Each node represents a lexical concept, attestation, or reference, with colors distinguishing their types. 

````python
import matplotlib.pyplot as plt
import networkx as nx
from rdflib import Graph, Namespace, URIRef, Literal, RDF, XSD
import textwrap

# --- Namespaces ---
EX = Namespace("http://example.org/")
RDF_NS = Namespace("http://www.w3.org/1999/02/22-rdf-syntax-ns#")
ONTOLEX = Namespace("http://www.w3.org/ns/lemon/ontolex#")
SN = Namespace("http://lila-erc.eu/data/lexicalResources/LatinWordNet/id/LexicalConcept/")
FRAC = Namespace("http://www.w3.org/ns/lemon/frac#")

# --- Create RDF graph ---
g = Graph()
g.bind("", EX)
g.bind("rdf", RDF_NS)
g.bind("ontolex", ONTOLEX)
g.bind("sn", SN)
g.bind("frac", FRAC)

lexical_concept = SN["02013448-v"]
attestation = EX["abituros_attestation"]
locus = URIRef("http://lila-erc.eu/data/corpora/Lasla/id/corpus/PlautusAmphitruo/Plautus_Amphitruo_PlAmphi.BPN_t_0001460")
observed_in = URIRef("https://lila-erc.eu/lodview/data/corpora/Lasla/id/corpus/PlautusAmphitruo/CiteStructure/Versus_208")

g.add((lexical_concept, RDF.type, ONTOLEX.LexicalConcept))
g.add((lexical_concept, FRAC.attestation, attestation))
g.add((attestation, RDF.type, FRAC.Attestation))
g.add((attestation, RDF_NS.value, Literal(
    "tradere si quae asportassent reddere se exercitum extemplo domum redducturum abituros agro Argiuos pacem atque otium dare illis sin aliter sient"
)))
g.add((attestation, FRAC.locus, locus))
g.add((attestation, FRAC.observedIn, observed_in))

# --- Build NetworkX graph ---
G = nx.DiGraph()
for s, p, o in g:
    G.add_node(str(s))
    G.add_node(str(o))
    G.add_edge(str(s), str(o), label=str(p).split("#")[-1])

# --- Node type colors ---
node_types = {}
for n in G.nodes:
    if n == str(lexical_concept):
        node_types[n] = "LexicalConcept"
    elif n == str(attestation):
        node_types[n] = "Attestation"
    elif n == str(locus) or n == str(observed_in):
        node_types[n] = "Reference"
    else:
        node_types[n] = "Other"

color_map = {
    "LexicalConcept": "#c60006",
    "Attestation": "#9b82d2",
    "Reference": "#f5B90f",
    "Other": "lightgray"
}
node_colors = [color_map[node_types[n]] for n in G.nodes]

# --- Layout ---
pos = nx.spring_layout(G, seed=42)

# --- Wrap labels ---
def wrap_label(label, width=20):
    # Display full URI for locus, observed_in, and lexical concept
    if label in [str(locus), str(observed_in), str(lexical_concept)]:
        return "\n".join(textwrap.wrap(label, width=30))
    # Otherwise, shorten and wrap
    if "/" in label:
        label = label.split("/")[-1]
    return "\n".join(textwrap.wrap(label, width=20))

labels = {n: wrap_label(n) for n in G.nodes}

# --- Draw nodes and edges ---
plt.figure(figsize=(12, 6))  # slightly bigger figure
nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=1800, edgecolors="black")
nx.draw_networkx_labels(G, pos, labels=labels, font_color="black", font_size=12)
nx.draw_networkx_edges(G, pos, arrowstyle="->", arrowsize=15, width=2)
edge_labels = {(u, v): d["label"] for u, v, d in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color="purple", font_size=12)

plt.axis("off")
plt.tight_layout()
plt.savefig("prototype_circular_nodes_with_uris.png", dpi=300)
plt.show()
````

After adding nouns, I obtain `all_token_URIs.csv`.

I then generatea `.ttl` using FrAC.

````python
import pandas as pd

# Load CSV
df = pd.read_csv("all_token_URIs.csv")

# Remove rows where synset_URIs is blank or NaN
df = df[df["synset_URIs"].notna() & (df["synset_URIs"].str.strip() != "")]

# Dataset URI
dataset_uri = "http://lila-erc.eu/data/dataset/PREMOVE/dataset"

# Now generate Turtle
with open("all_token_URIs.ttl", "w", encoding="utf-8") as f:
    # Write Turtle prefixes
    f.write("""@prefix : <http://lila-erc.eu/data/dataset/PREMOVE/id/attestation/> . 
@prefix rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#> .
@prefix ontolex: <http://www.w3.org/ns/lemon/ontolex#> .
@prefix sn: <http://lila-erc.eu/data/lexicalResources/LatinWordNet/id/LexicalConcept/> .
@prefix frac: <http://www.w3.org/ns/lemon/frac#> .
@prefix dcterms: <http://purl.org/dc/terms/> .

""")

    for _, row in df.iterrows():
        synset_uris = row["synset_URIs"]
        synset_id = synset_uris.replace("http://lila-erc.eu/data/lexicalResources/LatinWordNet/id/LexicalConcept/", "")
        token = str(row["token"])
        attestation = f"{token}_attestation"
        
        if pd.notna(row["SENTENCE"]):
            sentence = str(row["SENTENCE"]).replace('"', '\\"').replace("\n", " ").replace("\r", " ")
            sentence = " ".join(sentence.split())
        else:
            sentence = ""
        
        token_uris = str(row["token_URIs"]).split(";") if pd.notna(row["token_URIs"]) else []
        passage_uri = str(row["passage_URIs"]) if pd.notna(row["passage_URIs"]) else ""

        f.write(f"sn:{synset_id} a ontolex:LexicalConcept ;\n")
        f.write(f"    frac:attestation :{attestation} .\n\n")

        f.write(f":{attestation} a frac:Attestation ;\n")
        f.write(f"    rdf:value \"{sentence}\" ;\n")
        if token_uris:
            locus_entries = ", ".join(f"<{uri.strip()}>" for uri in token_uris if uri.strip())
            f.write(f"    frac:locus {locus_entries} ;\n")
        f.write(f"    frac:observedIn <{passage_uri}> ;\n")
        f.write(f"    dcterms:isPartOf <{dataset_uri}> .\n\n")

print("‚úÖ Done! Turtle file generated as 'all_token_URIs.ttl'")
````

Representing the SPARQL Query for WordNet Synsets.
 
```sparql
PREFIX frac: <http://www.w3.org/ns/lemon/frac#>
PREFIX ontolex: <http://www.w3.org/ns/lemon/ontolex#>
PREFIX dcterms: <http://purl.org/dc/terms/>
PREFIX lila: <http://lila-erc.eu/ontologies/lila/>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>

SELECT ?attestation ?token ?passage ?text
WHERE {
  VALUES ?syn { <http://lila-erc.eu/data/lexicalResources/LatinWordNet/id/LexicalConcept/02013448-v> }
    ?syn frac:attestation ?attestation .
    ?attestation rdf:value ?text .
    ?attestation frac:locus ?token ;
                 frac:observedIn ?passage .
}
ORDER BY ?token


````python
import networkx as nx
import matplotlib.pyplot as plt

# Define triple patterns extracted from the SPARQL WHERE clause
triples = [
    ("?syn", "frac:attestation", "?attestation"),
    ("?attestation", "rdf:value", "?text"),
    ("?attestation", "frac:locus", "?token"),
    ("?attestation", "frac:observedIn", "?passage")
]

# Create a directed graph
G = nx.DiGraph()

# Add edges and store predicates as labels
for s, p, o in triples:
    G.add_edge(s, o, label=p)

# Layout: increase k for more spacing
pos = nx.spring_layout(G, seed=42, k=3.5, iterations=300)

if "?text" in pos:
    pos["?text"][1] += 0.4

if "?token" in pos:
    pos["?token"][0] += 0.5

if "?passage" in pos:
    pos["?passage"][0] -= 0.5
    
# Prepare node colors (blue for variables, green for IRIs/constants)
node_colors = ["#a3c4f3" if n.startswith("?") else "#90ee90" for n in G.nodes()]
labels = {n: n for n in G.nodes()}

# Plot setup
plt.figure(figsize=(10, 5))
nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=2000, edgecolors="black")
nx.draw_networkx_labels(G, pos, labels=labels, font_color="black", font_size=14, font_weight="bold")
nx.draw_networkx_edges(G, pos, arrowstyle="->", arrowsize=15, width=1, connectionstyle="arc3,rad=0.05")

# Edge labels (predicates)
edge_labels = {(u, v): d["label"] for u, v, d in G.edges(data=True)}
nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels=edge_labels,
    font_color="purple",
    font_size=14,
    label_pos=0.5,
    rotate=False
)

#plt.title("SPARQL WHERE Subgraph: LatinWordNet Attestation Query", fontsize=14, fontweight="bold")
plt.axis("on")
plt.tight_layout()
plt.savefig("sparql_subgraph_latinwordnet.jpeg", dpi=300, bbox_inches="tight", transparent=True)
plt.show()
````

Representing the SPARQL Query for VerbNet Classes.

```sparql
PREFIX frac: <http://www.w3.org/ns/lemon/frac#>
PREFIX ontolex: <http://www.w3.org/ns/lemon/ontolex#>
PREFIX lila: <http://lila-erc.eu/ontologies/lila/>
PREFIX vn: <http://premon.fbk.eu/resource/>
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>

SELECT DISTINCT ?lemmaLabel ?vnClass ?attestation ?text
WHERE {
  VALUES ?lemmaLabel { "concurro" }  

  ?lemma a lila:Lemma ;
         rdfs:label ?lemmaLabel .

  ?token lila:hasLemma ?lemma .
  ?attestation frac:locus ?token .

  ?vnClass frac:attestation ?attestation .
  FILTER(STRSTARTS(STR(?vnClass), "http://premon.fbk.eu/resource/"))
}
ORDER BY ?vnClass ?text


````python
import networkx as nx
import matplotlib.pyplot as plt

# Triple patterns extracted from the SPARQL WHERE clause
triples = [
    ("?lemma", "rdf:type", "lila:Lemma"),
    ("?lemma", "rdfs:label", "?lemmaLabel"),
    ("?token", "lila:hasLemma", "?lemma"),
    ("?attestation", "frac:locus", "?token"),
    ("?vnClass", "frac:attestation", "?attestation")
]

# Create a directed graph
G = nx.DiGraph()

# Add edges and store predicates as labels
for s, p, o in triples:
    G.add_edge(s, o, label=p)

# Layout: increase k to push nodes farther apart
pos = nx.spring_layout(G, seed=42, k=3.5, iterations=300)

if "lila:Lemma" in pos:
    pos["lila:Lemma"][0] -= 0.5

if "?lemmaLabel" in pos:
    pos["?lemmaLabel"][0] += 0.5

# Prepare node colors and labels
node_colors = ["#a3c4f3" if n.startswith("?") else "#90ee90" for n in G.nodes()]
labels = {n: n for n in G.nodes()}

# Draw figure
plt.figure(figsize=(10, 5))
nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=2000, edgecolors="black")
nx.draw_networkx_labels(G, pos, labels=labels, font_color="black", font_size=14, font_weight="bold")
nx.draw_networkx_edges(G, pos, arrowstyle="->", arrowsize=15, width=1, connectionstyle="arc3,rad=0.05")

# Edge labels (predicates)
edge_labels = {(u, v): d["label"] for u, v, d in G.edges(data=True)}
nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels=edge_labels,
    font_color="purple",
    font_size=14,
    label_pos=0.5,
    rotate=False
)

#plt.title("SPARQL WHERE Subgraph: 'concurro'", fontsize=14, fontweight="bold")
plt.axis("on")
plt.tight_layout()
plt.savefig("sparql_subgraph_verbnet.jpeg", dpi=300, bbox_inches="tight", transparent=True)
plt.show()
````
